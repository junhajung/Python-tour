```python
import csv
import pandas as pd
import folium
import matplotlib.pyplot as plt
import seaborn as sns

# read
file = open('csv/4_문화생활.csv', 'r', encoding='cp949')
rd = csv.reader(file)

df = pd.DataFrame(rd)

# column 0번째 행 이름으로 설정
df.rename(columns = df.iloc[0], inplace=True)

# 0번째 행 column명으로 설정하였으니 0번째 행 삭제
df = df.drop(df.index[0])

# 열 삭제
df = df.drop(['총 저축 평균 잔액', '총 대출 평균 잔액', '서적문구 평균 소비금액'], axis=1)


# object타입 -> int 타입으로 변경
df['스포츠 평균 소비금액'] = pd.to_numeric(df['스포츠 평균 소비금액'])
df['영화공연 평균 소비금액'] = pd.to_numeric(df['영화공연 평균 소비금액'])
df['오락 평균 소비금액'] = pd.to_numeric(df['오락 평균 소비금액'])


# 공백이 들어가면 plot 그릴 때, 에러가 나서 공백을 제거한 이름으로 컬럼명 재설정
df.rename(columns={"스포츠 평균 소비금액": "스포츠"}, inplace=True)
df.rename(columns={"영화공연 평균 소비금액": "영화공연"}, inplace=True)
df.rename(columns={"오락 평균 소비금액": "오락"}, inplace=True)

# 연령대라는 column으로 각각 그룹화하여 평균을 구한 값을 dataframe으로 생성
df = df.groupby(['연령대'], as_index=False).mean()
df

# 한글처리
plt.rc("font", family="Malgun Gothic")
sns.set(font="Malgun Gothic", 
        rc = {'figure.figsize':(25,6)},
        style='darkgrid')

plt.title('연령별 여가 소비금액')
plt.xlabel('연령대')
plt.ylabel('금액')


plt.plot(df['연령대'], df.스포츠, marker='s', color='r', label='스포츠')
plt.plot(df['연령대'], df.영화공연, marker='s', color='g', label='영화공연')
plt.plot(df['연령대'], df.오락, marker='s', color='b', label='오락')

plt.legend()
plt.show()

```
